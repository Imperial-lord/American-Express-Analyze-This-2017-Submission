### PCA

```python
from sklearn.decomposition import PCA as sklearnPCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.datasets.samples_generator import make_blobs

from pandas.tools.plotting import parallel_coordinates
```

```python
X = dataYear.copy()
X.drop('target',inplace=True,axis=1)
y = dataYear['target']
```

```python
y_ = target_
```

```python
X_norm = (X - X.min())/(X.max() - X.min())
```

```python
pca = sklearnPCA(n_components=2) #2-dimensional PCA
transformed = pd.DataFrame(pca.fit_transform(X_norm))
```

```python
plt.scatter(transformed[y==0][0], transformed[y==0][1], label='Class 0', c='red')
plt.scatter(transformed[y==1][0], transformed[y==1][1], label='Class 1', c='blue')

plt.legend()
plt.show()
```

### LDA

```python
lda = LDA(n_components=2) #2-dimensional LDA
lda_transformed = pd.DataFrame(lda.fit_transform(X_norm, y_))
```

```python
plt.scatter(lda_transformed[y==0][0], lda_transformed[y==0][1], label='Class 0', c='red')
plt.scatter(lda_transformed[y==1][0], lda_transformed[y==1][1], label='Class 1', c='blue')
plt.legend()
plt.show()
```
